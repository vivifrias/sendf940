##Estudio



1.- El participante recibe una figura compleja (en este caso, la Figura de Rey-Osterrieth) que consta de 18 elementos.
Se le pide al participante que copie la figura tan precisamente como sea posible en una hoja en blanco.
Cada uno de los 18 elementos de la figura se puntúa individualmente según la precisión en la reproducción.


2.- Después de un intervalo de tres minutos, se le pide al participante que dibuje nuevamente la figura de memoria.
La puntuación se realiza de manera similar a la tarea de copia, evaluando la precisión en la reproducción de cada elemento.
Sistema de Puntuación:

3.- Cada elemento de la figura puede recibir una puntuación máxima de 2 puntos si se dibuja y coloca correctamente.
Se asignan puntajes menores para dibujos correctos pero colocados incorrectamente, dibujos incorrectos pero reconocibles y colocados correctamente, dibujos incorrectos pero reconocibles y colocados incorrectamente, y 0 puntos si está ausente o no reconocible.
Los puntajes individuales se suman para obtener un puntaje total, con un puntaje máximo de 36 para la figura completa.
Análisis de Resultados:

4.- Los resultados se pueden analizar de varias maneras, incluyendo el uso de percentiles para comparar el rendimiento del individuo con poblaciones normales.
En este estudio específico, se encontraron diferencias estadísticamente significativas entre pacientes con esclerosis múltiple y sujetos de control, destacando las mayores discrepancias en la tarea de copia y en la reproducción de memoria.

### <span style="color:blue">Descripción del problema</span>

Las enfermedades degenerativas del sistema nervioso son prevalentes en nuestra sociedad y afectan a millones de personas en todo el mundo. Ya se han realizado estudios en los que se les asigna a personas con estas enfermedades degenerativas, como la artritis o la esclerosis múltiple, la tarea de dibujar la figura de Rey-Osterreith, primero copiándola y luego reproduciéndola de memoria, con el fin de identificar estas enfermedades degenerativas. Logran esto viendo qué tan precisamente pueden recrear la imagen dada. Nuestro objetivo es combinar el uso de un sensor giroscópico y un sensor de presión para observar la diferencia en los resultados entre sujetos sanos y sujetos con artritis o esclerosis múltiple. Planeamos hacerlo creando un dispositivo de escritura, como un bolígrafo, que contenga estos sensores y pidiendo a los sujetos que completen la figura de Rey-Osterreith. Al combinar los datos de velocidad angular (obtenidos del giroscopio), presión aplicada (obtenida de los sensores de presión), tiempo necesario para completar la figura y precisión en la recreación, esperamos poder identificar a las personas afectadas por la artritis o la esclerosis múltiple.



La investigación "Automated Scoring of a Neuropsychological Test: The Rey-Osterrieth Complex Figure" presenta un método automatizado para evaluar el rendimiento en el test de la Figura Compleja de Rey-Osterrieth (FCRO). El FCRO es una prueba neuropsicológica que evalúa la función cognitiva, incluida la memoria visual, la atención y la organización.

El método automatizado propuesto utiliza un sistema de visión artificial para capturar y analizar los dibujos de los participantes. El sistema identifica y mide los componentes del dibujo, como las líneas, los ángulos y los círculos. A partir de esta información, el sistema calcula una serie de puntuaciones que se utilizan para evaluar el rendimiento del participante.

Esto fue probado en una muestra de 100 participantes, incluidos pacientes con trastornos neurológicos y personas sanas. Los resultados mostraron que el método automatizado era capaz de proporcionar puntuaciones precisas y consistentes. Además, el método fue capaz de detectar diferencias significativas en el rendimiento entre los participantes con y sin trastornos neurológicos.

Esta propuesta tiene el potencial de mejorar la evaluación del FCRO. El método es más rápido y eficiente que la evaluación manual, y es más objetivo y reproducible. Además, el método puede ser utilizado para realizar evaluaciones remotas, lo que puede facilitar el acceso a la evaluación neuropsicológica.

3-El sistema identifica y mide los componentes del dibujo.
4-El sistema calcula una serie de puntuaciones que se utilizan para evaluar el rendimiento del participante.

Estos resultados de la investigación sugieren que el método automatizado es una herramienta valiosa para la evaluación del FCRO. El método es preciso, eficiente y objetivo, y tiene el potencial de mejorar la evaluación de la función cognitiva.

#### Lectura del sensor de presión y configuración del tiempo

HAL_RTC_GetDate y HAL_RTC_GetTime obtienen la fecha y la hora actual del módulo Reloj en Tiempo Real que luego se convierten en una cadena. Usamos el conversor analógico-digital para consultar un valor del sensor que hemos conectado a través de nuestro STM y lo covnertimos. Entonces eso también es formateado en una cadena para que podamos ver estos resultados en nuestra terminal Hercules. Estas consultas ocurren aproximadamente cada tres segundos.
